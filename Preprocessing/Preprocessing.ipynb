```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
SRC_PATH = "./selwood.txt"
DEST_PATH = "./"
```

```python
def normalize(df):
    result = df.copy()
    for c in df.columns[1:]:
        max_value = df[c].max()
        min_value = df[c].min()
        result[c] = (df[c] - min_value) / (max_value - min_value)
    return result
```

```python
df_training = pd.DataFrame()
df_test = pd.DataFrame()
```

```python
df = pd.read_csv(SRC_PATH)
df
```

```python
df_norm = normalize(df)
df_norm
```

```python
# Randomly sample 70% of dataframe for training set, rest 30% for test
df_training = df_norm.sample(frac=0.7, random_state=2)
df_test = df_norm.loc[~df_norm.index.isin(df_training.index)]
    
df_test.to_csv(DEST_PATH + "test.csv", index=False, header=None)
df_training.to_csv(DEST_PATH + "training.csv", index=False, header=None)
```
